```python
import random
import pandas as pd

# Kamus leetspeak sederhana untuk mengganti huruf
leet_dict = {
    'a': ['4', '@'],
    'b': ['8'],
    'c': ['(', '<'],
    'd': ['d'],
    'e': ['3'],
    'g': ['9'],
    'h': ['h'],
    'i': ['1', '!', 'i'],
    'l': ['1'],
    'm': ['m'],
    'n': ['n'],
    'o': ['0'],
    'p': ['p'],
    'r': ['r'],
    's': ['5', '$'],
    't': ['7'],
    'u': ['u'],
    'v': ['v'],
    'w': ['w'],
    'x': ['x'],
    'y': ['y'],
    'z': ['2']
}

# Fungsi untuk mengubah kalimat menjadi gaya leetspeak
def to_leetspeak(text):
    result = ""
    for char in text:
        if char.lower() in leet_dict and random.random() < 0.7:
            result += random.choice(leet_dict[char.lower()])
        elif char == ' ' and random.random() < 0.5:
            result += random.choice([" ", "'", '"'])
        else:
            result += char
    return result

# Fungsi untuk membuat fake link pendek
def generate_fake_short_link():
    code = ''.join(random.choices('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789', k=6))
    return f"tinydd.co/{code}"
```

```python
# Daftar frasa acak bertema spam dengan variasi konten
spam_templates = [
    "buruan main di situs terbaik {site}",
    "deposit cuma {amount} bisa jadi {bonus}",
    "auto maxwin tiap hari, cuma di {site}",
    "web slot terpercaya, gampang wd",
    "promo gila! deposit {amount} bonus {bonus}",
    "langsung gas maxwin bareng kami di {site}",
    "link slot gacor hari ini: {link}",
    "situs anti rungkad, buktiin skrg!",
    "dapetin jackpot dengan modal receh",
    "main aman, menang mudah di {site}",
    "claim bonus tanpa syarat, klik {link}",
    "event hadiah total ratusan juta!",
    "modal kecil cuan besar, ayo daftar!",
    "akses game hoki via {site}",
    "top up hari ini dapat cashback {percent}%",
    "hanya hari ini, maxwin menanti kamu",
    "cuan harian dari game paling seru",
    "slot gampang menang, langsung spin",
    "main santai, saldo naik sendiri",
    "terbukti gacor! link: {link}"
]

# Fungsi untuk membuat kalimat random dengan leetspeak
def generate_custom_spam(): 
    message = []
    for _ in range(random.randint(3, 5)):
        template = random.choice(spam_templates)
        # Isian acak
        site = to_leetspeak(random.choice(["situsraja", "gacorgaming", "mantapbet", "zonacuan", "maxslot"]))
        link = generate_fake_short_link()
        amount = str(random.choice([25, 50, 100, 200, 500]))
        bonus = str(int(amount) * 2)
        percent = str(random.choice([10, 25, 50, 75, 100]))
        sentence = template.format(site=site, link=link, amount=amount, bonus=bonus, percent=percent)
        sentence = to_leetspeak(sentence)
        message.append(sentence)
    return "\n".join(message)

# Generate data dummy spam 
custom_spam_messages = []
random.seed(42)
for _ in range(20):
    msg = generate_custom_spam()
    custom_spam_messages.append({"pesan": msg, "label": "spam"})

df_custom_spam = pd.DataFrame(custom_spam_messages)
df_custom_spam
```
